```python
from langchain_community.document_loaders import WhatsAppChatLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Load chats
docs = WhatsAppChatLoader('WhatsApp_Chat_with_Mom.txt').load()

# Join all content into one string
chat_text = "\n".join([doc.page_content for doc in docs])

chat_text

chat_splitter = RecursiveCharacterTextSplitter(chunk_size=200,separators=["\n\n"])
splitted_chats = chat_splitter.split_text(chat_text)

splitted_chats

```
